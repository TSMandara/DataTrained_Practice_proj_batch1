```python
try:
    dataset_name_a=driver.find_elements_by_xpath("//tr/td/table/tbody/tr/td[2]/p/b/a")  #Extracting Dataset name
    for k in dataset_name_a:
        dataset_name.append(k.text)
except NoSuchElementException:   #Handling Exceptions
    dataset_name.append('--')
except NoSuchElementException:
    dataset_name.append('--')
try:
    task_a=driver.find_elements_by_xpath("//td[3]/p[@class='normal']")    #Extracting Task
    for r in task_a:
        task.append(r.text)
except NoSuchElementException:   #Handling Exceptions
    task.append('--')
except NoSuchElementException:
    task.append('--')

try:
    attribute_types_a=driver.find_elements_by_xpath("//td[4]/p[@class='normal']") #Extracting Attribute types
    for e in attribute_types_a:
        attribute_types.append(e.text)
except NoSuchElementException:   #Handling Exceptions
    attribute_types.append('--')
except NoSuchElementException:
    attribute_types.append('--')
    
try:
    no_of_instances_a=driver.find_elements_by_xpath("//td[5]/p[@class='normal']")   #Extracting No. of instances
    for v in no_of_instances_a:
        no_of_instances.append(v.text)
except NoSuchElementException:     #Handling Exceptions
    no_of_instances.append('--')
except NoSuchElementException:
    no_of_instances.append('--')

try:
    no_of_attributes_a=driver.find_elements_by_xpath("//td[6]/p[@class='normal']")   #Extracting No. of attributes
    for m in no_of_attributes_a:
        no_of_attributes.append(m.text)
except NoSuchElementException:     #Handling Exceptions
    no_of_attributes.append('--')
except NoSuchElementException:
    no_of_attributes.append('--')

try:
    year_a=driver.find_elements_by_xpath("//td[7]/p[@class='normal']")    #Extracting Year
    for t in year_a:
        year.append(t.text)
except NoSuchElementException:   #Handling Exceptions
    year.append('--')
except NoSuchElementException:
    year.append('--')
```
